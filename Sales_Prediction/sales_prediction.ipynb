```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv(r"c:\users\DELL\Downloads\advertising.csv")
data.head()
```

```python
features=data[['TV','Radio','Newspaper']]
target=data['Sales']
```

```python
from sklearn.model_selection import train_test_split
X_train_data,X_test_data,y_train_data,y_test_data=train_test_split(features,target,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression 
model=LinearRegression() 
model.fit(X_train_data,y_train_data)
```

```python
predicted_sales=model.predict(X_test_data)
```

```python
from sklearn.metrics import mean_squared_error
mse=mean_squared_error(y_test_data,predicted_sales)
print("Mean Squared Error:",mse)
```

```python
print("model trained to predict sales based on advertising spend and performance is evaluated using mse ")
print("this show how advertising budgets influence sales")
```

```python
plt.figure()
plt.scatter(y_test_data,predicted_sales)
plt.xlabel("actual sales")
plt.ylabel("predicted sales")
plt.title("actual vs predicted sales")
plt.show()
```
